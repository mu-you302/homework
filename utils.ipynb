```python
## Analyze the image width and height

import os
import cv2
from matplotlib import pyplot as plt

# analyze the image W, H
dataset_path = "./cell_images"
W, H = [], []

for dir in os.listdir(dataset_path):
    for file in os.listdir(os.path.join(dataset_path, dir)):
        if not file.endswith('.png'):   # non-image
            continue
        img = cv2.imread(os.path.join(dataset_path, dir, file))
        h, w, _ = img.shape
        W.append(w) # width
        H.append(h)
```

```python
# plot
plt.figure(figsize=(12, 5))
plt.subplot(1, 3, 1)
plt.hist(W, bins=100, alpha=0.5, label='Width')
plt.xlabel('Width', fontsize=20)
plt.ylabel('Frequency', fontsize=20)

plt.subplot(1, 3, 2)    
plt.hist(H, bins=100, alpha=0.5, label='Height')
plt.xlabel('Height', fontsize=20)
# plt.ylabel('Frequency', fontsize=20)

# disparity of the W, H
plt.subplot(1, 3, 3)
diff = [abs(W[i]-H[i]) for i in range(len(W))] 
plt.hist(diff, bins=100, alpha=0.5, label='Disparity')
plt.xlabel('Disparity', fontsize=20)
# plt.ylabel('Frequency', fontsize=20)
plt.savefig('figures/size_analyze.png', dpi=600)
plt.show()
```
